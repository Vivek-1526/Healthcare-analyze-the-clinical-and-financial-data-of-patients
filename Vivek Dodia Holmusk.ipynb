```python
import numpy as np
import pandas as pd 
import matplotlib as plt
```

```python
df1=pd.read_csv('bill_amount.csv')
df2=pd.read_csv('bill_id.csv')
df3=pd.read_csv('clinical_data.csv',parse_dates=["date_of_admission","date_of_discharge"])
df4=pd.read_csv('demographics.csv')
```

```python
df1
```

```python
df12 = pd.merge(df1, df2 ,on="bill_id")
```

```python
df12
```

```python
df123 = pd.merge(df12, df3 ,left_on="patient_id",right_on="id")
```

```python
df123
```

```python
df_final = pd.merge(df123, df4 ,left_on="patient_id",right_on="patient_id")
```

```python
df_final.replace(to_replace ="m",
                 value ="Male",inplace=True)
df_final.replace(to_replace ="f",
                 value ="Female",inplace=True)
```

```python
df_final.columns
```

```python
df_final.groupby(['gender'], as_index=True)['patient_id'].count().plot(rot=90,kind="bar")
```

```python
df_final.groupby(['race'], as_index=True)['patient_id'].count().plot(rot=90,kind="bar")
```

```python
df_final.groupby(['resident_status'], as_index=True)['patient_id'].count().plot(rot=90,kind="bar")
```
